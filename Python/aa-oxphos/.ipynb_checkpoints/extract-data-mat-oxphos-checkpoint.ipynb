```python
import scipy.io
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
```

```python
folder = 'D:\\Data analysis\\Mitography\\AA-OXPHOS\\all_oxphos_results-exp145'
files = os.listdir(folder)
files = [file for file in files if file.endswith('.mat')]
```

```python
exp_all = []
ctaa_all = []
cell_all = []
area_all = []
length_all = []
width_all = []
ar_all = []
oxphosparam_all = []

colnames = ['exp', 'ctaa', 'cell', 'area', 'length', 'width', 'ar', 'oxphosparam']
# put all mito in one dataframe
data_df_all = pd.DataFrame(columns=colnames)

filenums = range(len(files))
for filenum in filenums:
    data = scipy.io.loadmat(os.path.join(folder,files[filenum]),simplify_cells=True)
    area = data['mitoinfo']['area']
    length = data['mitoinfo']['length']
    width = data['mitoinfo']['width']
    ar = data['mitoinfo']['ar']
    oxphosparam = data['mitoinfo']['oxphosparam']
    exp = np.repeat(data['mitoinfo']['exp'], len(area))
    ctaa = np.repeat(data['mitoinfo']['ctaa'], len(area))
    cell = np.repeat(filenum+1, len(area))
    exp_all.append(exp)
    ctaa_all.append(ctaa)
    area_all.append(area)
    length_all.append(length)
    width_all.append(width)
    ar_all.append(ar)
    oxphosparam_all.append(oxphosparam)
    
    # create dataframe for cell with all info per mitos
    data_df = pd.DataFrame(np.vstack((exp, ctaa, cell, area, length, width, ar, oxphosparam)).T, 
                           columns=colnames)
    # concat mito to all mito dataframe
    data_df_all = pd.concat([data_df_all,data_df])
# change datatypes from str to numeric (float or int) for all numeric data columns in dataframe for all mitos
data_df_all['area'] = pd.to_numeric(data_df_all['area'], downcast="float")
data_df_all['length'] = pd.to_numeric(data_df_all['length'], downcast="float")
data_df_all['width'] = pd.to_numeric(data_df_all['width'], downcast="float")
data_df_all['ar'] = pd.to_numeric(data_df_all['ar'], downcast="float")
data_df_all['oxphosparam'] = pd.to_numeric(data_df_all['oxphosparam'])
data_df_all['cell'] = pd.to_numeric(data_df_all['cell'])
```

```python
#### all
mito_all = data_df_all
savename = 'mito_all_exp145.csv'
mito_all.to_csv(os.path.join(folder,savename), index=False)

#### ct
# all detected mitos
ct_mito_all = data_df_all[data_df_all['ctaa'] == 'ct']
savename = 'ct_mito_all_exp145.csv'
ct_mito_all.to_csv(os.path.join(folder,savename), index=False)
# all MDMs
ct_mdm = ct_mito_all[ct_mito_all['area'] < 0.086]
savename = 'ct_mdm_exp145.csv'
ct_mdm.to_csv(os.path.join(folder,savename), index=False)
# all mitos > 0.086 µm2
ct_mito = ct_mito_all[ct_mito_all['area'] > 0.086]
savename = 'ct_mito_exp145.csv'
ct_mito.to_csv(os.path.join(folder,savename), index=False)

# all oxphos+ all mitos
ct_mito_all_oxpp = ct_mito_all[ct_mito_all['oxphosparam'] == 1]
savename = 'ct_mito_all_oxpp_exp145.csv'
ct_mito_all_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all oxphos- all mitos
ct_mito_all_oxpn = ct_mito_all[ct_mito_all['oxphosparam'] == 0]
savename = 'ct_mito_all_oxpn_exp145.csv'
ct_mito_all_oxpn.to_csv(os.path.join(folder,savename), index=False)
# all oxphos+ mdms
ct_mdm_oxpp = ct_mdm[ct_mdm['oxphosparam'] == 1]
savename = 'ct_mdm_oxpp_exp145.csv'
ct_mdm_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all oxphos- mdms
ct_mdm_oxpn = ct_mdm[ct_mdm['oxphosparam'] == 0]
savename = 'ct_mdm_oxpn_exp145.csv'
ct_mdm_oxpn.to_csv(os.path.join(folder,savename), index=False)
# all oxphos+ mitos
ct_mito_oxpp = ct_mito[ct_mito['oxphosparam'] == 1]
savename = 'ct_mito_oxpp_exp145.csv'
ct_mito_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all oxphos- mitos
ct_mito_oxpn = ct_mito[ct_mito['oxphosparam'] == 0]
savename = 'ct_mito_oxpn_exp145.csv'
ct_mito_oxpn.to_csv(os.path.join(folder,savename), index=False)

#### aa
# all detected mitos
aa_mito_all = data_df_all[data_df_all['ctaa'] == 'aa']
savename = 'aa_mito_all_exp145.csv'
aa_mito_all.to_csv(os.path.join(folder,savename), index=False)
# all MDMs
aa_mdm = aa_mito_all[aa_mito_all['area'] < 0.086]
savename = 'aa_mdm_exp145.csv'
aa_mdm.to_csv(os.path.join(folder,savename), index=False)
# all mitos > 0.086 µm2
aa_mito = aa_mito_all[aa_mito_all['area'] > 0.086]
savename = 'aa_mito_exp145.csv'
aa_mito.to_csv(os.path.join(folder,savename), index=False)

# all oxphos+ all mitos
aa_mito_all_oxpp = aa_mito_all[aa_mito_all['oxphosparam'] == 1]
savename = 'aa_mito_all_oxpp_exp145.csv'
aa_mito_all_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all oxphos- all mitos
aa_mito_all_oxpn = aa_mito_all[aa_mito_all['oxphosparam'] == 0]
savename = 'aa_mito_all_oxpn_exp145.csv'
aa_mito_all_oxpn.to_csv(os.path.join(folder,savename), index=False)
# all oxphos+ mdms
aa_mdm_oxpp = aa_mdm[aa_mdm['oxphosparam'] == 1]
savename = 'aa_mdm_oxpp_exp145.csv'
aa_mdm_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all oxphos- mdms
aa_mdm_oxpn = aa_mdm[aa_mdm['oxphosparam'] == 0]
savename = 'aa_mdm_oxpn_exp145.csv'
aa_mdm_oxpn.to_csv(os.path.join(folder,savename), index=False)
# all oxphos+ mitos
aa_mito_oxpp = aa_mito[aa_mito['oxphosparam'] == 1]
savename = 'aa_mito_oxpp_exp145.csv'
aa_mito_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all oxphos- mitos
aa_mito_oxpn = aa_mito[aa_mito['oxphosparam'] == 0]
savename = 'aa_mito_oxpn_exp145.csv'
aa_mito_oxpn.to_csv(os.path.join(folder,savename), index=False)
```
