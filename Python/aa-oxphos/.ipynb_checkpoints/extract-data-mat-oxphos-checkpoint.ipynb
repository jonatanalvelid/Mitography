```python
import scipy.io
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
```

```python
folder = 'D:\\Data analysis\\Mitography\\AA-PEX-OXPHOS\\analysis_exp2'
files = os.listdir(folder)
files_oxp = [file for file in files if file.endswith('.mat') and file.startswith('oxphos')]
files_pex = [file for file in files if file.endswith('.mat') and file.startswith('pex')]
files = [file for file in files if file.endswith('.mat')]
print(files_oxp)
print(files_pex)
print(files)
```

```python
exp_all = []
ctaa_all = []
cell_all = []
area_all = []
length_all = []
width_all = []
ar_all = []
oxphosparam_all = []
pexparam_all = []

colnames = ['exp', 'ctaa', 'cell', 'area', 'length', 'width', 'ar', 'oxphosparam']
colnames_pex = ['exp', 'ctaa', 'cell', 'area', 'pexparam']
# put all mito in one dataframe
data_df_all = pd.DataFrame(columns=colnames)
data_df_pex_all = pd.DataFrame(columns=colnames_pex)

filenums = range(len(files))
for filenum in filenums:
    if files[filenum].startswith('oxphos'):
        data = scipy.io.loadmat(os.path.join(folder,files[filenum]),simplify_cells=True)
        area = data['mitoinfo']['area']
        length = data['mitoinfo']['length']
        width = data['mitoinfo']['width']
        ar = data['mitoinfo']['ar']
        oxphosparam = data['mitoinfo']['oxphosparam']
        exp = np.repeat(data['mitoinfo']['exp'], len(area))
        ctaa = np.repeat(data['mitoinfo']['ctaa'], len(area))
        cell = np.repeat(filenum+1, len(area))
        exp_all.append(exp)
        ctaa_all.append(ctaa)
        area_all.append(area)
        length_all.append(length)
        width_all.append(width)
        ar_all.append(ar)
        oxphosparam_all.append(oxphosparam)

        # create dataframe for cell with all info per mitos
        data_df = pd.DataFrame(np.vstack((exp, ctaa, cell, area, length, width, ar, oxphosparam)).T, 
                               columns=colnames)
        # concat mito to all mito dataframe
        data_df_all = pd.concat([data_df_all,data_df])
        
    elif files[filenum].startswith('pex'):
        data = scipy.io.loadmat(os.path.join(folder,files[filenum]),simplify_cells=True)
        pexparam = data['mitoinfo']['pexparam']
        area = data['mitoinfo']['area']
        exp = np.repeat(data['mitoinfo']['exp'], len(pexparam))
        ctaa = np.repeat(data['mitoinfo']['ctaa'], len(pexparam))
        cell = np.repeat((filenum % 4) + 1, len(pexparam))
        pexparam_all.append(pexparam)
        
        # create dataframe for cell with pex info per mitos
        data_df_pex = pd.DataFrame(np.vstack((exp, ctaa, cell, area, pexparam)).T, 
                               columns=colnames_pex)
        # concat pex to all pex dataframe
        data_df_pex_all = pd.concat([data_df_pex_all, data_df_pex])
        
# concat all pex info to all mito df with oxphos info
data_df_all.insert(np.shape(data_df_all)[1]-1, 'pexparam', data_df_pex_all['pexparam'].astype('int64'))

# change datatypes from str to numeric (float or int) for all numeric data columns in dataframe for all mitos
data_df_all['area'] = pd.to_numeric(data_df_all['area'], downcast="float")
data_df_all['length'] = pd.to_numeric(data_df_all['length'], downcast="float")
data_df_all['width'] = pd.to_numeric(data_df_all['width'], downcast="float")
data_df_all['ar'] = pd.to_numeric(data_df_all['ar'], downcast="float")
data_df_all['oxphosparam'] = pd.to_numeric(data_df_all['oxphosparam'])
data_df_all['cell'] = pd.to_numeric(data_df_all['cell'])
```

```python
#### ct
# all detected mitos
ct_mito_all = data_df_all[data_df_all['ctaa'] == 'ct']
savename = 'ct_mito_all_exp2.csv'
ct_mito_all.to_csv(os.path.join(folder,savename), index=False)
# all MDMs
ct_mdm = ct_mito_all[ct_mito_all['area'] < 0.086]
savename = 'ct_mdm_exp2.csv'
ct_mdm.to_csv(os.path.join(folder,savename), index=False)

# all oxphos+ all mitos
ct_mito_all_oxpp = ct_mito_all[ct_mito_all['oxphosparam'] == 1]
savename = 'ct_mito_all_oxpp_exp2.csv'
ct_mito_all_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all oxphos- all mitos
ct_mito_all_oxpn = ct_mito_all[ct_mito_all['oxphosparam'] == 0]
savename = 'ct_mito_all_oxpn_exp2.csv'
ct_mito_all_oxpn.to_csv(os.path.join(folder,savename), index=False)

# all pex+ all mitos
ct_mito_all_pexp = ct_mito_all[ct_mito_all['pexparam'] == 1]
savename = 'ct_mito_all_pexp_exp2.csv'
ct_mito_all_pexp.to_csv(os.path.join(folder,savename), index=False)
# all pex- all mitos
ct_mito_all_pexn = ct_mito_all[ct_mito_all['pexparam'] == 0]
savename = 'ct_mito_all_pexn_exp2.csv'
ct_mito_all_pexn.to_csv(os.path.join(folder,savename), index=False)

# all pex+ oxphos+ all mitos
ct_mito_all_pexp_oxpp = ct_mito_all[(ct_mito_all['pexparam'] == 1) & (ct_mito_all['oxphosparam'] == 1)]
savename = 'ct_mito_all_pexp_oxpp_exp2.csv'
ct_mito_all_pexp_oxpp.to_csv(os.path.join(folder,savename), index=False)
# all pex- oxphos- all mitos
ct_mito_all_pexn_oxpn = ct_mito_all[(ct_mito_all['pexparam'] == 0) & (ct_mito_all['oxphosparam'] == 0)]
savename = 'ct_mito_all_pexn_oxpn_exp2.csv'
ct_mito_all_pexn_oxpn.to_csv(os.path.join(folder,savename), index=False)
```
